Empezaremos haciendo limpieza de las respuestas de los m2 de jardin:

```python
#Vamos a ver que registros no son numÃ©ricos

registros_no_numericos = df['m2_jardin'][df['m2_jardin'].apply(lambda x: not str(x).isdigit())]
registros_no_numericos.value_counts()
```

Mediante una transformaciÃ³n, convertiremos todos los registros a numÃ©ricos, de la siguiente manera:
- Si el registro es un numero se queda como estÃ¡.
- Si el registro contiene un nÃºmero pero tiene otros caracteres (ejemplo: 200m), se queda solo el nÃºmero
- Si el registro es solo texto se transformarÃ¡ a 0

```python
#Primero necesitamos convertir la columna a str
df['m2_jardin'] = df['m2_jardin'].astype(str)
```

```python
#Nos ayudamos de expresiones regulares
import re
def transformar_registro(registro):
    if registro.isdigit():  
        return int(registro)  
    else:
        numero_encontrado = re.search(r'\d+', registro)  
        if numero_encontrado:  
            return int(numero_encontrado.group())  
        else:
            return 0  
```

```python
#hacemos la transformaciÃ³n correspondiente
df['m2_jardin'] = df['m2_jardin'].apply(transformar_registro)
```
