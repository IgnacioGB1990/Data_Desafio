```python
#Borramos columnas que no necesitamos

df.drop(columns= ['#', 'Start Date (UTC)', 'Submit Date (UTC)', 'Network ID', 'Tags'], inplace= True)
```

```python
df.columns= ['Edad', 'Individuos_Hogar', 'Provincia', 'Tema_Interés', 'Tiempo_Ducha', 'Frecuencia_lavadora',
                    'Frecuencia_lavavajillas', 'Modo_ECO', 'Reciclar_agua', 'Reciclar_material', 'Consumo_carne',
                     'km_coche', 'Gasto_mascotas', 'm2_jardin', 'Consumo_agua_€']
```

```python
df.head()
```

Para variables categóricas:

```python
def representacion (columna):
    print(df[columna].value_counts(dropna= False))
    
    df[columna].value_counts().plot(kind='bar', ylabel= None, title= str(columna))
    plt.show()
```
